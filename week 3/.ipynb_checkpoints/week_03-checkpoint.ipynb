```python
from matplotlib import pyplot as plt
import numpy as np

def simulate_ddm(D,f,z):
    
    # Initialize variables
    i = 0
    N_correct = 0
    N_wrong = 0
    
    delta = []
    delta.append(N_correct - N_wrong)
    
    running = True
    while running:
        
        for j in range(D):
            
            dot_direction = np.random.binomial(1, f)
            N_correct += dot_direction
            N_wrong += (1-dot_direction)
        
        i += 1
        delta.append(N_correct-N_wrong)
        
        if abs(delta[i]) > z:
            running = False
        
    DT = i - 1
    accuracy = delta[i]> 0
    
    return delta[:-1], DT, accuracy
```

```python
D = 10
f = 0.55
z = 60
```

```python
# Plot different deltas
fig = plt.figure()
ax1 = fig.add_subplot(111)
for i in range(10):
    delta, DT, accuracy = simulate_ddm(D,f,z)
    l = ax1.plot(delta,'-',linewidth = 3)

x = np.linspace(-10,100,100)
y_pos = (np.ones(x.shape) *  z)
y_neg = (np.ones(x.shape) * -z)
ax1.plot(x,y_pos,'--')
ax1.plot(x,y_neg,'--')
ax1.set_xlim(-5,100)
ax1.set_ylim(-80,80)
ax1.set_xlabel('Number of Frames')
ax1.set_ylabel('Accumulation of Evidence')
ax1.set_title('Diffusion Decision Model')
```

```python
# Plot histogram of reaction times
fig = plt.figure()
ax2 = fig.add_subplot(111)
DT_list = []

for i in range(1000):
    delta, DT, accuracy = simulate_ddm(D,f,z)
    DT_list.append(DT)

ax2.hist(DT_list,histtype ='stepfilled',bins = 20)
ax2.set_xlabel('Response Time')
ax2.set_ylabel('Occurence')
ax2.set_title('Diffusion Decision Model')
```

```python
# Plot speed-accuracy tradeoff

z_vals = [2, 4, 8, 16, 32]
accuracy_list = []
DT_list = []
mean_accuracies = []
mean_DT = []

for ii in range(len(z_vals)):
    for jj in range(100):
        delta, DT, accuracy = simulate_ddm(D,f,z_vals[ii])
        DT_list.append(DT)
        accuracy_list.append(accuracy)
    
    mean_accuracies.append(np.mean(accuracy_list))
    mean_DT.append(np.mean(DT_list))

        
fig = plt.figure()
ax1 = fig.add_subplot(131)
ax2 = fig.add_subplot(132)
ax3 = fig.add_subplot(133)


ax1.plot(z_vals,mean_accuracies)
ax1.set_xlabel('Values of Threshold')
ax1.set_ylabel('Accuracy')

ax2.plot(z_vals,mean_DT)
ax2.set_xlabel('Values of Threshold')
ax2.set_ylabel('Response Time')

ax3.plot(mean_DT,mean_accuracies)
ax3.set_xlabel('Response Time')
ax3.set_ylabel('Accuracy')
```
